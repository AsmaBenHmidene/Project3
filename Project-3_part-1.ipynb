```python
import pandas as pd

url1="https://datasets.imdbws.com/title.basics.tsv.gz"
url2="https://datasets.imdbws.com/title.ratings.tsv.gz"
url3="https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
basics = pd.read_csv(url1, sep='\t', low_memory=False)
basics.head(1)
```

```python
ratings = pd.read_csv(url2, sep='\t', low_memory=False)
ratings.head(1)
```

```python
akas = pd.read_csv(url3, sep='\t', low_memory=False)
akas.head(1)
```

```python
import numpy as np
basics = basics.replace({'\\N':np.nan})
```

```python
basics.head(5)
```

```python
basics['runtimeMinutes'].isna().sum()
```

```python
basics['genres'].isna().sum()
```

```python
basics.info()
```

```python
#Eliminate movies that are null for runtimeMinutes, Eliminate movies that are null for genre
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
```

```python
basics.head(5)
```

```python
basics['runtimeMinutes'].isna().sum()
```

```python
basics['genres'].isna().sum()
```

```python
basics['titleType'].value_counts()
```

```python
basics['startYear'].value_counts()
```

```python
basics['startYear'].isna().sum()
```

```python
basics = basics.dropna(subset=['startYear'])
```

```python
basics_filteryear = (basics['startYear'] >='2000')  &  (basics['startYear'] <='2021')
basics_filteryear.value_counts()
```

```python
basics_filtertype = basics['titleType']=='movie'
basics_filtertype.value_counts()
```

```python
basics = basics.loc[basics_filtertype & basics_filteryear, :]
basics
```

```python
basics['startYear'].value_counts()
```

```python
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
basics
```

```python
basics['genres'].value_counts()
```

```python
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```
